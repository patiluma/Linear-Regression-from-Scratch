## Linear Regression from Scratch

```python
import numpy as np
```

### LinearRegression class with 4 methods:
- fit()     → learns weights and bias from data
- predict() → uses learned weights to make predictions
- mse()     → measures average error
- r2()      → measures how good the model is

```python
class LinearRegression:

  def __init__(self, learning_rate = 0.01, n_iterations = 1000):
    self.learning_rate = learning_rate
    self.n_iterations = n_iterations
    self.weights = None
    self.bias = None

  def fit(self, X, y):
    n_samples, n_features = X.shape
    self.weights = np.zeros(n_features)
    self.bias = 0

    for _ in range(self.n_iterations):
      y_pred = np.dot(X,self.weights) + self.bias

      dw = (1/n_samples) * np.dot(X.T,(y_pred - y))
      db = (1/n_samples) * np.sum(y_pred-y)

      self.weights -= self.learning_rate * dw
      self.bias -= self.learning_rate * db

  def predict(self,X):
    return np.dot(X,self.weights) + self.bias

  def mse(self,X, y):
    return np.mean((self.predict(X) - y)**2)

  def r2(self, X, y):
    sse = np.sum((y - self.predict(X))**2)
    tot = np.sum((y - np.mean(y))**2)
    return 1 - sse/tot
```

###Creating Data

```python
np.random.seed(42)

X = np.random.uniform(0,10, size = (100, 1))
y = 3.5 * X.ravel() + 7 + np.random.normal(0,3,100)

model = LinearRegression(learning_rate = 0.01, n_iterations = 1000)
model.fit(X,y)
```

###Results

```python
print(f"Weight: {model.weights[0]:.4f}")
print(f"Bias: {model.bias:.4f}")
print(f"MSE: {model.mse(X,y):.4f}")
print(f"R2: {model.r2(X,y):.4f}")
```
